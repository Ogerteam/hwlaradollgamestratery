### Rules of the Game

There is a field of 32 vases, and some known number of them are marked as golden ones, how many golden vases there are depends of the level of Scroll of Midas. The player has archeolory tools looking like pickaxe. Player uses its limited number of pickaxe, to open vases one by one, looking for the big grand prize called Lara's Doll. At the vases there are 2 other lesser grand prizes. Any golden vase contains 5 times more awards from a normal vase - for example golden one containing a Lara prize would give you 5 Lara prizes. When the player accidently opened any of the grand prizes, player can refresh the field populating it with another 32 vases, or may contunue selecting vases from the current field. The game finishes when pickaxes finishes. 
Origin of the game is taken from Hero-Wars RPG[1].

### Aim of the player
Is to get maximum number of Lara Dolls using a fixed number of pickaxes.

### Why it is actually a game?

Because when player hits a lesser grand prize, the player has to choose, what to do?  To open all the rest from the field, or to hit refresh button and reload another 32 vases. At that moment how to take decision? That this document explains.

# Strategy

1. Start opening one by one vases, starting with the golden ones.

2. When the golden are finished continue with the rest of non-golden vases.

3. If at any time you hit Lara Doll grand prize, refresh and start again the game.

4. If you hit grand prize, that is the moment to get decision whether to refresh or continue.

5. Whether to continue or to refresh depends on how many vases are left unopened at the field, and as well as from the level of Scroll of Midas. 

6. Consult the table below, according to the number of unopened vases, and level of Scroll of Midas, the table gives you what to choose to refresh or to continue, according to the higher value of mathematical expectation for each decision. Mathematical expectation as known as expected value of the yield. At the only decision point it selects the move with better expected yield. 

## Why that could work

Imagine we have only 32 pickaxes. We have started to opening vases first golden ones and next continue with the rest. Let's after opening first $k$ vases we hit lesser grand prize and we become able to reset the field. The question is, which is better for the remainder of the pickaxes $r = 32 - k$, to continue or refresh? 

The approach is to compare the expected yields - the mathematical expectation for both possible decisions. Decision one is to choose to stay and open all remaining, expected value in that case is 1 if all golden vases are already opened. Decision two is choose to refresh and to invest the remaining pickaxes $r = 32 - k$ into opening new golden vases from a new field, because there are vases with higher yield that might be better. The table just compares the two expected values, and gives which one has bigger expected value for yield.

## The Decision Table 

In the table below, first column $k$ is the number of already opened vases, the second column $r$ is the number of remaining vases, and all the rest of the colums are designated as $SM 1$, $SM 2$, ... $SM 14$, are corresponding to the initial number of golden vases per field - the level of Scroll of Midas. 

1. "YES" means if you have a chance to refresh, then refresh - because mathematical expectation to open the rest in a new field could give you more.

2. "NO" means if you have a chance to refresh, do not - because mathematical expectation to open the rest in a new field could give you less.

3. "Y/N" means if you have a chance to refresh, it does not matter - because mathematical expectation to open the rest in a new field or stay are equal.

If the difference between expected value for refreshing and staying is positive it recommends to refresh, if it is negative it recommends to stay, if it is equal it says it does not matter.

```python
n = 32
golden_prize = 5
common_prize = 1
```

#### Probability to find Lara in first $k$ vases from $n$ available vases
She could be in one of $k$ boxes from $n$ available. The probability space is consisted of $n$ elements $n$ possible independent locations of Lara, from them only $k$ are golden ones and are chosen. The probability to be in one of them is, or in other words the sum of probabilities of the elementary events:
$$
    P = \frac{k}{n}
$$

```python
def probFindingLara(k, n) :
  """
  Probability for finding Lara in first k opened vases from total of n vases.
  Args:
    k: - number of vases to be opened
    n: - total vase number
  Returns: Probability
  """
  return k/n
```

```python
def expectedValue_OpenFromBeginning(k, g, n):
  """
  Mathematical Expectation, average yield, if planned to open k number of vases from the beginning.
  Args:
    k: - number of vases to be opened, starting from golden ones from the beginning of the field.
    g: - number of golden vases, expectation by 5
    n: - number of all vases on the field
  Returns: Mathematical Expectation, Expected Value
  """

  ''' if looking into just golden ones '''
  if ( k <= g ) :
    return golden_prize*probFindingLara(k, n);

  ''' if looking into more just golden ones '''

  ''' probability to find Lara in golden vases '''
  probInGolden = probFindingLara(g, n);

  ''' prob to find Lara in common vase after you dig all golden - relative probability, remained k-g tools, and remained n-g vases.'''
  relprobInCommon = probFindingLara(k-g, n-g);
  probInCommon = (1-probInGolden) * relprobInCommon

  ''' mathematical expectation '''
  return golden_prize*probInGolden + common_prize*probInCommon
```

```python
def expectedValue_OpenTheRest(r, g, n) :
  """
  Mathematical Expectation, average yield, if k vases are already opened, and remained only r, and you decide to open the rest r of the vases.
  Args:
    r: - number of vases remained for opening.
    g: - number of golden vases, expectation by 5
    n: - number of all vases on the field
  Returns: Mathematical Expectation, Expected Value
  """
  ''' already opened '''
  k = n-r
  ''' probability to be in golden remaining ones '''  
  p = 0
  ''' if still unopened golden vases, better open them, '''
  if ( k < g ) :
    ''' probability to find Lara in the remaining golden vases, it remained g-k golden in the field of r vases. '''
    p = probFindingLara(g-k, r)
  return common_prize*(1-p) + golden_prize*p;  
```

## Decision Table

Decision table, already opened $k$ vases, and $r=32-k$ remained, what to do open rest $r$ or refresh and get fresh $r$ vases. The value under each labeled with the Scroll of Midas level column, is the difference, is how the expectation is changed if refreshed.

It just compares which one from the two choices has better expectation value. Table prints the difference - how much would be changed the current expectation if the player refreshes the field.


```python
def printDecisionTable(b) :
    ''' b - boolean flags only '''
    header   =   "|  k |  r |"
    hlinesep =   "+----+----+"
    for g in range(1,15):
        header = header + f" SM {g:3} |"
        hlinesep = hlinesep + "--------+"
    
    print(hlinesep)
    print ("|  Decision table, already opened k vases, r=32-k remained, what to do, to open the rest r or refresh and get fresh k vases?")
    print ("|  The value under each labeled with the Scroll of Midas level column, is the difference, is how the expectation is changed if refreshed.")
    print(hlinesep)
    print(header)
    print(hlinesep)
    
    ''' k - opened already '''
    ''' r - remained ''' 
    
    for k in range(0,n):
        r = n-k 
        line = f"| {k:2} | {r:2} |";
        for g in range(1,15):
            rest = expectedValue_OpenTheRest(r,g,n)
            newbegin = expectedValue_OpenFromBeginning(r,g,n)
            m = newbegin-rest
            if (m>0):
                if (b):
                    line += "  YES   |"
                else: 
                    line += f" +{m:.3f} |"
            else:
                if (m==0) :
                    if (b) :
                        line += "  Y/N   |"
                    else:    
                        line += f"  {m:.3f} |"
                else:
                    if (b):
                       line += "  NO    |" 
                    else:
                       line += f" {m:.3f} |"

        print(f"{line}")
    print(hlinesep)
```
